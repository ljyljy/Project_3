```python
import pandas as pd
import requests
from datetime import datetime
import matplotlib.pyplot as plt
import sys
sys.path.append('./Resources')
from config import dkey
```

```python
prop_2016 = pd.read_csv('./Resources/properties_2016.csv')
prop_2017 = pd.read_csv('./Resources/properties_2017.csv')
```

```python
location_group_2016 = prop_2016[['regionidcity', 'latitude', 'longitude']].groupby('regionidcity').mean()
location_group_2016 = location_group_2016.reset_index()
location_group_2016['latitude'] /= 1e6
location_group_2016['longitude'] /= 1e6
```

```python
location_group_2017 = prop_2017[['regionidcity', 'latitude', 'longitude']].groupby('regionidcity').mean()
location_group_2017 = location_group_2017.reset_index()
location_group_2017['latitude'] /= 1e6
location_group_2017['longitude'] /= 1e6
```

```python
Jan_1_2016 = 1451610000
Feb_1_2016 = 1454288400
Aug_1_2016 = 1470013200
Sep_1_2016 = 1472691600
Jan_1_2017 = 1483232400
Feb_1_2017 = 1485910800
Aug_1_2017 = 1501549200
Sep_1_2017 = 1504227600
one_day = 86400
```

```python
df_filled_2016 = pd.DataFrame()
```

```python
for i in location_group_2016.index:
    unix_time = Jan_1_2016
    while unix_time <= Feb_1_2016:
        url = ('https://api.darksky.net/forecast/{3}/{0},{1},{2}?extend=hourly')\
            .format(location_group_2016['latitude'][i], location_group_2016['longitude'][i], int(unix_time), dkey)
        temp_data = requests.get(url).json()
        temp_data = temp_data['hourly']['data']
        df_temp = pd.DataFrame(temp_data)
        df_temp['site_id'] = location_group_2016['regionidcity'][i]
        df_temp.time = pd.to_datetime(df_temp['time'],unit='s')
        df_filled_2016 = pd.concat([df_filled_2016, df_temp], ignore_index=True, sort = 'False')
        unix_time += one_day
        if i % 20 == 0:
            print(datetime.utcfromtimestamp(unix_time))
            print(i)
            print(df_temp.shape)
            print('*' * 35)
    unix_time = Aug_1_2016
    while unix_time <= Sep_1_2016:
        url = ('https://api.darksky.net/forecast/{3}/{0},{1},{2}?extend=hourly')\
            .format(location_group_2016['latitude'][i], location_group_2016['longitude'][i], int(unix_time), dkey)
        temp_data = requests.get(url).json()
        temp_data = temp_data['hourly']['data']
        df_temp = pd.DataFrame(temp_data)
        df_temp['site_id'] = location_group_2016['regionidcity'][i]
        df_temp.time = pd.to_datetime(df_temp['time'],unit='s')
        df_filled_2016 = pd.concat([df_filled_2016, df_temp], ignore_index=True, sort = 'False')
        unix_time += one_day
        if i % 20 == 0:
            print(datetime.utcfromtimestamp(unix_time))
            print(i)
            print(df_temp.shape)
            print('*' * 35)
    
```

```python
df_filled_2016.to_csv('./Resources/weather_initial_2016.csv', index  = False)
```
