```python
import pandas as pd
import numpy as np
```

```python
df_2016 = pd.read_csv('./Resources/train_2016_v2.csv')
df_2017 = pd.read_csv('./Resources/train_2017.csv')
df_prop_2016 = pd.read_csv('./output/prop_2016_refilled.csv')
df_prop_2017 = pd.read_csv('./output/prop_2017_refilled.csv')
```

```python
df_2016 = df_2016.merge(df_prop_2016, on = 'parcelid', how = 'left')
df_2017 = df_2017.merge(df_prop_2017, on = 'parcelid', how = 'left')
```

```python
df = pd.concat([df_2016, df_2017])
```

```python
df = df.reset_index(drop = True)
```

```python
columns = {'buildingqualitytypeid_fill': 'buildingqualitytypeid', 'lotsizesquarefeet_refill': 'lotsizesquarefeet',
          'propertycountylandusecode_fill': 'propertycountylandusecode', 'regionidcity_fill': 'regionidcity', 
          'regionidneighborhood_fill': 'regionidneighborhood', 'regionidzip_fill': 'regionidzip', 
          'unitcnt_fill': 'unitcnt', 'yearbuilt_fill':'yearbuilt'}
df = df.drop_duplicates(subset = ['parcelid', 'logerror'])
df.drop(['lotsizesquarefeet'], axis = 1, inplace = True)
final_list = ['buildingqualitytypeid', 'lotsizesquarefeet', 'propertycountylandusecode', 'regionidcity', 
             'regionidneighborhood', 'regionidzip', 'unitcnt', 'yearbuilt',
             'latitude', 'longitude', 'logerror']
df = df.rename(columns = columns)[final_list]
df.to_csv('./output/clean_final.csv', index = False)
```
